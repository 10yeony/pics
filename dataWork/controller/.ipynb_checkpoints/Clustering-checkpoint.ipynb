```python
import numpy as np #판다스 친구 넘파이
import pandas as pd # 판다스
from pandas import DataFrame as df
import re # 정규식 사용을 위함
import nltk
import csv
from konlpy.tag import Okt
from konlpy.utils import pprint
from collections import Counter
```

```python
import os
dataset = pd.read_csv("total_dataset_trim.csv",encoding='utf-8')
dataset
```

```python
# 복합명사로 이루어진 태그는 잘라서 길이 2~7 사이의 단어만 추출
def getTagList(value):
    okt=Okt()
    parseTagList=[]
    tempTagList=[]
    tags=(value).split('#') # 태그별로 자르기
    for tag in tags:
        if len(tag)>1:
            okt.nouns(tag)  # 1개 태그 내 복합 명사 자르기
            for j in okt.nouns(tag):
                if(len(j)>=1|len(j)<7):
                    tempTagList.append(j)
    parseTagList.append(tempTagList)
    return parseTagList
```

```python
def getCoreTags(n):
    coreTagList=[]
    okt=Okt()    
    for i in range(n):
        coreTag=[]
        desc=[i for i in okt.nouns(dataset.description[i]) if len(i)>=1 ]
        tags=(dataset.tag[i]).split('#')
        tags=getTagList(dataset.tag[i])  # tag를 한번 더 파싱
        for t in tags:
            if t in desc:
                coreTag.append(t)
        if len(dataset.tag[i])<=3:
            coreTagList.append(desc)
        elif len(coreTag)>3:
            coreTagList.append(coreTag)
        else:
            coreTagList.append(tags)
    return coreTagList
```
